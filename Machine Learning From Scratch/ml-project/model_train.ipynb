```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer,make_column_selector as selector
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import KNNImputer,SimpleImputer,IterativeImputer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder,RobustScaler,StandardScaler

train_cat=Pipeline([
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder(handle_unknown='ignore'))
])

train_num=Pipeline([
    ('imputer',KNNImputer(n_neighbors=5)),
    ('scaler',RobustScaler())
])
preprocessor=ColumnTransformer([
    ('cat',train_cat,selector(dtype_include=['category','object'])),
    ('num',train_num,selector(dtype_include=['number']))
])


```

```python
from xgboost import XGBRFClassifier
from catboost import CatBoostClassifier
from sklearn.linear_model import SGDClassifier
model=Pipeline([
    ('preprocessing',preprocessor),
    ('model',SGDClassifier(loss='log_loss',max_iter=100,tol=1e-2,learning_rate='adaptive'))
])
model
```
